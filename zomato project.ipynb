# 6] conver the data type of column - rate

```python
def handleRate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')[0]
        return float(value)

dataframe['rate']=dataframe['rate'].apply(handleRate)
dataframe.head()
```

```python
dataframe.info()
```

# Q1 Which locations have the most restaurants?

```python
dataframe['location'].value_counts().head(10)
```

# Bar Chart

```python
top_locations=dataframe['location'].value_counts().head(10)

top_locations.plot(kind='bar')
plt.title('Top 10 Locations with Most Restaurants')
plt.xlabel('location')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45)
plt.show()
```

we analyzed the location column to identify areas with the highest number of restaurants. 
Using value counts,we found the top 10 locations where restaurants are most concentrated.
This helps understand popular food hubs and business-dense areas in the city.

# Insight

 BTM has the highest number of retaurants,making it the most dominant food hub in the dataset
 Locations such as HSR and Koramangala also show a high concentration of restaurants,indicating strong customer demand and commercial food activity.
 In contrast,areas like Bellandur have fewer restaurant listings,suggesting lower density.

# Q2 Which mode(online or offline) has received the maximum rating ?

# Box plot Chart

```python
dataframe.head()
```

```python
dataframe.boxplot(column='rate',by='online_order')
plt.title('Average Rating: Online vs offline')
plt.suptitle('')
plt.xlabel('Online Order')
plt.ylabel('Rating')
plt.show()
```

Restaurants that offer online ordering have a higher median rating compared to offline-only restaurants.
The box plot also shows fewer low-rating outliers for online-order restaurants,indicating better and more consistent customer satisfaction.

```python
dataframe.groupby('online_order')['rate'].mean()
```

# Q3 which rating most restaurants receive?

```python
dataframe.head()
```

```python
most_common_rating = dataframe['rate'].mode()[0]
print(f"The most common rating is:{most_common_rating}")
```

```python
rating_counts =dataframe['rate'].value_counts().sort_index()

rating_counts.plot(kind='bar',figsize=(10,5),color='lightgreen')
plt.title('Number of retaurants by rating')
plt.xlabel('rating')
plt.ylabel('number of restaurant')
plt.xticks(rotation=0)
plt.show()
```

# The majority of restaurants have received a rating of 3.9
